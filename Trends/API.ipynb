```python
import os
import quandl
import datetime
import wget
import hvplot.pandas
import requests
import json
import pandas as pd
from pandas.tseries.offsets import MonthEnd
%matplotlib inline
```

```python
house_sale = pd.DataFrame(quandl.get("YALE/RHPI"))
```

```python
house_sale.index.name = "Time"
house_sale.head()
```

```python
house_sale.hvplot(
    y = "Index",
    xlabel="Date",
    ylabel="Price Index",
    title ="House Price Index",
    rot=90, 
    height = 500,
    width = 600
)
```

```python
redfin_sale=pd.read_csv("med_sale_price_data.csv")
redfin_sale["Time"]=pd.to_datetime(redfin_sale['Month of Period End'])+MonthEnd(1)
redfin_sale.drop(["Month of Period End","Region"],axis=1,inplace=True)
redfin_sale.set_index("Time")
redfin_sale = redfin_sale[["Time","Avg. Median Sale Price"]]
redfin_sale.head()
```

```python
redfin_sale.hvplot(
    x = "Time",
    y = "Avg. Median Sale Price",
    xlabel="Date",
    ylabel="Sale Price",
    title ="Redfin Avg. Median Sale Price",
    rot=90, 
    height = 500,
    width = 600
)
```

```python
zillow_url = "https://files.zillowstatic.com/research/public_csvs/median_sale_price/Metro_median_sale_price_uc_sfrcondo_sm_month.csv?t=1626991465"
zillow_file = wget.download(zillow_url)
```

```python
zillow_sale = pd.read_csv(zillow_file)
zillow_sale.tail()
```

```python
zillow_sale.drop(zillow_sale.index[1:],0, inplace=True)
zillow_sale.head()
```

```python
zillow_sale.drop(zillow_sale.columns[:5],axis=1, inplace=True)
zillow_sale.head()
```

```python
zillow_sale = zillow_sale.transpose()
```

```python
zillow_sale.index.name = "Time"
zillow_sale.set_axis(["Sale Price"],axis=1,inplace = True)
zillow_sale.head()
```

```python
zillow_sale.hvplot(
    xlabel="Time",
    ylabel="Sale Price",
    title ="Zillow Median Sale Price",
    rot=45, 
    height = 500,
    width = 1000,
    
).opts(yformatter='%.0f')
```
